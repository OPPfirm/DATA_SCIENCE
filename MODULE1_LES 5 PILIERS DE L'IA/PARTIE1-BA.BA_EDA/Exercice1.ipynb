### 2. Analysez la variable "Country".

```python
#appliquer value_count() pour avoir l'occurence
X=df["Country"].value_counts().sort_values() 
print(X) 


#appliquer  " .plot("kind"=bar) " a X pour visulaiser sur barchart ,l'occurence
plt.figure()
X.plot(kind="bar",rot=0)
plt.title("Occurence de chaque pays entre \n1970-2010")

```

### 7. Observez l'évolution des dépenses en santé de l'Allemagne et de la Grande Bretagne à travers le temps.

### 8. Observez l'histograme, la boxplot, et les statistiques de "Life_Expectancy" propre à chaque pays.

```python


#statistique 
print(df.groupby('Country')["Life_Expectancy"].describe())

#histogram
plt.figure(figsize=(10,8))
#sns.catplot(data=df, x="Country",y="Life_Expectancy",kind='bar')
sns.displot(data=df , x="Life_Expectancy",col="Country")

#boxplot
sns.catplot(data=df,y="Country",x="Life_Expectancy", kind="box" )
```

```python
# 9. Observez la relation entre la durée de vie et les dépenses dans la santé, aux États-Unis, de 2000 à 2020
new_df2= df[(df["Country"]=='USA') & (df["Year"] >= 2000) ]

plt.figure(figsize=(10,8))
#boolean indexing
#sns.scatterplot(data=new_df2 ,x="Life_Expectancy",y="Spending_USD")
#other methodd without boolean indexing but .query()
sns.scatterplot(data=df.query("Country=='USA' & Year >= 2000") ,y="Life_Expectancy",x="Spending_USD")
# Ajouter les annotations (années) sur chaque point
filtered_df = df.query("Country=='USA' & Year >= 2000")
for i, row in filtered_df.iterrows():
    plt.annotate(row["Year"], (row["Spending_USD"], row["Life_Expectancy"]), 
                 textcoords="offset points", xytext=(0, 5), ha='center', fontsize=9, color='black')

plt.show()
plt.show()
```
