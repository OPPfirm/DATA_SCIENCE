# Annalyse technique

## 1. Analysez la relation entre x et y.


```python
sns.relplot(data=df, x="x",y="y",hue="dataset",col="dataset")
plt.legend( title="dataset",bbox_to_anchor=(1.05,1) ,loc="upper left")

plt.show()
```

## 3. Analysez maintenant les statistiques de chaque "paquet". Que constatez-vous ?


### Analyse de la variable x

```python
plt.figure(figsize=(8,6))
#histogramme
fig,ax = plt.subplots(2,1,sharex=True)
plt.suptitle("Analyse de la varible x ")
sns.histplot(data=df ,x="x",ax=ax[0])
#boite a moustache
sns.boxplot(data=df, x="x",ax=ax[1])

plt.show()
```

### Analyse de la variable y

```python
fig,ax=plt.subplots(2,1,sharex=True)
plt.suptitle("Analyse de la variable y")
sns.histplot(data=df,x="y",ax=ax[0])
sns.boxplot(data=df,x="y",ax=ax[1])
plt.show()
```
