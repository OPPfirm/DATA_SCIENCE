```python
## Dataset de test 
np.random.seed(42)
n_samples =100
X_numeric= np.random.rand(n_samples,2)#2 variables continues (height,weight)
X_discrete=np.random.randint(1,10,size=(n_samples,2))#2 variables discretes (age,floor)
X_categorical=np.random.choice(['A', 'B','C'], size=(n_samples,1))#1 variable categorielle
y=np.random.rand(n_samples)*10 #variables cibles

#creation du dataframe
df=pd.DataFrame(np.hstack( (X_numeric,X_discrete,X_categorical) ), 
                        columns=["height","weight","age","floor","Category"])
                        
df["y"]=y#creation d'une nouvelle collone avec les datas de "y"

df.head()
```

```python
#definition des colonnes
categorical_cols_4= []
continuous_col=["height","weight"]
discret_col=["age","floor"]

#transformation des variables categorielles avec sélection de caracteristiques
cat_transformer= Pipeline(
    steps=[
        ("OneHot Encoder", OneHotEncoder(sparse_output=True)),
        ("SelectK Best",SelectKBest(score_func=mutual_info_regression,k=2)),        
    ]
)

#definition du ColumnTransformer
column_transformer_4= ColumnTransformer(
    transformers=[
        ("catagories", cat_transformer,categorical_cols_4),
        ("standardizer",StandardScaler(),continuous_col),
        ("MinMax",MinMaxScaler(),discret_col)
    ],
)

#creation de la pipeline finale
pipeline_4 = Pipeline(
    steps=[
        ("column transformer" ,column_transformer_4),
        ("Polynomial Features", PolynomialFeatures(degree=4,interaction_only=True)),
        ("MinMax Scaler",MinMaxScaler())
    ]
)
pipeline_4
```

```python
x=df.drop(columns=["y"])
y=df["y"]
pd.DataFrame(pipeline_4.fit_transform(x,y)).head()
```
