### Data

```python
df=sns.load_dataset("diamonds")
df.head()
```

### Train Test Split

```python
train_set,test_set =train_test_split(df,test_size=0.2,random_state=0)
# random_state=n , est comme random.seed(n) qui permet de fixer toutes  les valeurs aleatoire produite pour la valeur de n 
print("train_set:",train_set.shape)
print("test_set:",test_set.shape)
```

## Train Set

### Encoding

```python
cut_order=['Fair','Good','Very Good','Premium','Ideal']
color_order=['J','I','H','G','F','E','D']
clarity_order = ['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1']

order=[cut_order,color_order,clarity_order]

encoder= OrdinalEncoder( categories=order )
encoding_results=encoder.fit_transform(train_set[["cut","color","clarity"]])
pd.DataFrame(encoding_results,columns=['cut', 'color', 'clarity']).head()
```

```python
ignore_warnings()

train_set.loc[:,['cut', 'color', 'clarity']] = encoding_results
train_set.head()
```

### Normalisation

```python
normaliser= MinMaxScaler()
train_set = pd.DataFrame( normaliser.fit_transform(train_set), columns=df.columns)
train_set.head()

```

### Transformation de test set


```python
encoding_results=encoder.transform(test_set[["cut","color","clarity"]])
pd.DataFrame(encoding_results).head()
```

```python
test_set.loc[:,['cut', 'color', 'clarity']] = encoding_results
test_set.head()
```

```python
test_set = pd.DataFrame( normaliser.transform(test_set), columns=test_set.columns )
test_set.head()
```

## Pipeline Composée

```python
from sklearn.compose import ColumnTransformer
```

```python
categorical_cols = liste_vd(data=df)
##liste_vd est ma propre fonction,elle ramene les variables discret de mon df en une liste  ici c'est : ['cut', 'color', 'clarity']
```

#### **note :**  ColumnTransformer() agit comme "Pipline" sauf qu'on peut y specifier les colonne sur les quelles on  veut traavailler (via 'transformer') et sur celle qu'on veut laisser tranquille (via 'remainder') 
#### - remainder : il prend deux valeurs 'drop' (supprime les autres du dataframe) et 'passthrough' (n'utilise pas les autres mais les conserves dans le dataset) 

```python
column_transformer=ColumnTransformer(transformers = [("encoder",OrdinalEncoder(),categorical_cols)] , remainder="passthrough")
column_transformer
```

```python
pipeline=make_pipeline(column_transformer,MinMaxScaler())
pipeline
```

```python
pd.DataFrame(pipeline.fit_transform(train_set),columns=df.columns ).head()
```

```python
pd.DataFrame(pipeline.transform(test_set), columns=df.columns).head()
```
