### **2- R2_SCORE**

```python
#€generation de data 
 

np.random.seed(0)
x=np.random.rand(20,1)*10
y=(3*x + np.random.randn(20,1) * 5).reshape(-1,1)

plt.scatter(x,y)
plt.show()
```

```python
model=LinearRegression()
model.fit(x,y)
y_pred= model.predict(x)

plt.scatter(x,y)
plt.plot(x,y_pred,c='black')
plt.show()
```

```python
for i in range(len(x)):
    plt.plot([x[i][0], x[i][0] ] , [y[i][0], y_pred[i][0]],color='red',linestyle='--',lw=2)
plt.scatter(x, y, s=50, c='black', zorder=2)
plt.plot(x,y_pred, color="green" , lw=3,label='Modele lineaire')
plt.show()
```

```python
r2 = r2_score(y, y_pred)
print(f"Le R² score est de: {r2}")

# Interpretation du R² score :
# Le R² score mesure la proportion de la variance de la variable dépendante (y) qui est expliquée par le modèle linéaire.
# Un R² de 0.7317 signifie que 73.17% de la variance de y est expliquée par le modèle.
# En d'autres termes, le modèle capture une part importante de la variabilité de la variable cible.
# Plus le R² est proche de 1, meilleur est l'ajustement du modèle aux données.
# Un R² de 0.7317 est généralement considéré comme un bon score, indiquant que le modèle a une capacité prédictive raisonnable.
```
