# Fonction Exponentielle

- Tracez la fonction exponentielle sur un intervalle [-10, 4]
- calculez exp(0)
- calculez exp(1)
- calculez exp(10000)

```python
X=np.linspace(-10,4,100)
y=np.exp(X)

plt.scatter(X, y, edgecolor='black', s=15)
plt.plot(X,y,color="red")
plt.title("fonction exponentielle")
plt.show()
```

# Fonction Sigmoid

- Rappelez la formule de la fonction sigmoid et écrivez la dans une fonction Python
- Calculez Sigmoid(0)
- Tracez la fonction sigmoid sur l'intervalle [-5, 5]
- Démontrez que les limites de la fonction sigmoid sont 0 et 1
- Calculez les valeurs de sigmoid(10000) sigmoid(-10000). Que se passe-t'il ?

```python
def sigmoid(x):
    return 1/(1+ np.exp(-x))
```

```python
sigmoid(0)
```

```python
X = np.linspace(-20, 20, 100)
y = sigmoid(X)

coord = 0
plt.scatter(X, y)
plt.vlines(x=coord, ymin=0, ymax=sigmoid(coord), color='r')
plt.hlines(y=sigmoid(coord), xmin=X.min(), xmax=coord, color='r')
plt.show()
```

```python
sigmoid(10000)
```

```python
sigmoid(-10000)
```

- Rappelez la formuler de la fonction logit et écrivez la dans une fonction Python

- Soit X = 2.24, y = sigmoig(X). z = logit(y). Que vaut z ?

```python
def logit(x):
    return np.log( x/(1-x) )
```

Le domaine de définition de la fonction Logit est `]0, 1[`

```python
X = np.linspace(-0.5, 1.5, 100)

y = X / (1 - X)

plt.plot(X, y)
plt.ylim(0)

plt.title("$X / (1 - X) > 0$")
plt.show()
```

```python
X = np.linspace(0.001, 0.999, 100)
y = logit(X)

plt.scatter(X, y)
plt.grid()
plt.show()
```
