# Les fonctions logarithmiques

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Quelques exemples de fonctions logarithmiques

```python
print("ln(2) = ", np.log(2)) # népérien
print("log10(100) = ", np.log10(100)) # décimale -> ordre de grandeur
print("log2(16) = ", np.log2(16)) # binaire -> théorie de l'information


print("log1p(1e-9) = ", np.log1p(1e-9)) # plus précis sur les calculs de petites valeurs X
print("log(1e-9) = ", np.log(1 + 1e-9))
```

## Comparaison de fonctions logarithmiques

```python
X = np.linspace(0.01, 10)
y = np.log(X)
y10 = np.log10(X)
y2 = np.log2(X)
y1p = np.log1p(X)

plt.scatter(X, y, label="log", s=5)
plt.scatter(X, y10, label="log10", s=5)
plt.scatter(X, y2, label="log2", s=5)
plt.scatter(X, y1p, label="log1p", s=5)

plt.grid()
plt.legend()
plt.show()
```
