# Introduction aux Polynômes

Un **polynôme** est une expression algébrique formée de la somme de plusieurs monômes. Un polynôme en une variable peut être écrit sous la forme :

$$ P(x) = a_nx^n + a_{n-1}x^{n-1} + \dots + a_1x + a_0 $$
    
où $a_i$ sont des coefficients réels (ou complexes), et $n$ est le degré du polynôme.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exemple : Polynôme de degré 2

```python
x = np.linspace(-10, 10, 400)
y2 = x**2 - 4*x + 3

plt.figure(figsize=(8, 5))
plt.plot(x, y2, label='Polynôme: $x^2 - 4x + 3$', color='blue')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.title('Polynôme de degré 2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

## Exemple : Polynôme de degré 3

```python
y3 = x**3 - 6*x**2 + 11*x - 6

plt.figure(figsize=(8, 5))
plt.plot(x, y3, label='Polynôme: $x^3 - 6x^2 + 11x - 6$', color='green')
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.title('Polynôme de degré 3')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

# Polynômes à Plusieurs variables (x, y).

Codons un exemple de polynome de a 2 variables, et 3 degrées : $f(x, y) = 0.5 x^3 + 3x^2 + 2xy + y^2 - 4x + 5y + 6$

```python
from mpl_toolkits.mplot3d import Axes3D

# Création des grilles pour x et y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Définition du polynôme à deux variables
Z = 0.5 * X**3 + 3*X**2 + 2*X*Y + Y**2 - 4*X + 5*Y + 6

# Tracé du graphe 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title("Polynôme de degré 2 à deux variables : $0.5 x^3 + 3x^2 + 2xy + y^2 - 4x + 5y + 6$")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```

## Nombre de termes d'un polynôme de degré $d$ à $n$ variables

La formule générale pour déterminer le **nombre de monômes distincts** (ou termes) d'un polynôme de degré $d$ à $n$ variables est donnée par :

$$ \text{Nombre de termes} = \binom{n + d}{d} $$

où $\binom{n + d}{d}$ est le coefficient binomial ("n plus d parmi d").

```python
from math import comb

def nombre_termes(d, n):
    return comb(n + d, d)


nombre_termes(3, 2)
```
