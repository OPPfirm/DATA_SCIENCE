# TP – Descente de Gradient et Convergence d'une suite 

Apres avoir étudié la fonction de `MSE` dans les précédents chapitres, et avoir calculé ses dérivées, il est temps de voir l'algorithme de la descente de gradient ! :)

Pour cela, partons d'une fonction simple :

$f(x) = (x-3)^2$

Nous allons ici étudier la convergende de la suite :

$x_{n+1}  = x_n - \eta f'(x_n) $

1. Calculer la dérivée de f
2. Identifier la condition de convergence de la suite
3. Trouver le point de convergence
4. Tracer la suite

```python
from sympy import symbols, diff
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Commencons par calculer la dérivée de f

```python
x = symbols('x')
f = (x - 3)**2

# Calcul de la dérivée symbolique
f_deriv = diff(f, x)
print(f"La dérivée de {f}, est {f_deriv}")
```

### 2. Étude de la suite

D'apres nos calculs, on cherche donc a étudier la suite :

$$x_{n+1}  = x_n - \eta (2 x_n - 6) $$

On peut simplifier et factoriser par $x_n$

$x_{n+1}  = x_n - \eta 2 x_n - 6 \eta $

$x_{n+1}  = x_n (1 - 2 \eta) - 6 \eta $

On a donc ici la formule d'une **suite affine** de la forme :

$$x_{n+1} = ax_n + b$$

#### Condition de convergence

Par définition, on sait que :
- si $|a| < 1$, alors la suite converge vers un point fixe
- sinon, elle diverge.


Ici, $a = (1 - 2 \eta)$

Donc, La suite converge si $|1 - 2 \eta| < 1$

Autrement dit, si $ -1 < 1 - 2 \eta < 1$

En simplifiant : $ 0 < \eta < 1$

En effet : 

```python
x0 = 100
eta = 0.8
n_iter = 20

f_prime = lambda x: 2 * (x - 3)

x_vals = [x0]

for n in range(n_iter):
    x_new = x_vals[-1] - eta * f_prime(x_vals[-1])
    x_vals.append(x_new)

x_vals
```

```python
x0 = 100
eta = 0.3
n_iter = 20

f_prime = lambda x: 2 * (x - 3)

x_vals = [x0]

for n in range(n_iter):
    x_new = x_vals[-1] - eta * f_prime(x_vals[-1])
    x_vals.append(x_new)

plt.plot(x_vals)
plt.title("0 < Eta < 1")
plt.show()
```

```python
x0 = 100
eta = 1.5
n_iter = 20

f_prime = lambda x: 2 * (x - 3)

x_vals = [x0]

for n in range(n_iter):
    x_new = x_vals[-1] - eta * f_prime(x_vals[-1])
    x_vals.append(x_new)

plt.plot(x_vals)
plt.title("Eta > 1")
plt.show()
```

### Point Fixe de la suite

Le point fixe, ou point de convergence, de la suite est donné par :

$$x = ax + b$$

(Quand cette formule est validée, la suite ne bouge plus)

En isolant x, on peut ainsi écrire :

$$x = \displaystyle \frac{b}{1 - a} = \displaystyle \frac{6 \eta}{1 - (1 - 2 \eta)} = \displaystyle \frac{6 \eta}{2 \eta } = 3 $$

```python
x0 = 10
eta = 0.3
n_iter = 20

f_prime = lambda x: 2 * (x - 3)

x_vals = [x0]

for n in range(n_iter):
    x_new = x_vals[-1] - eta * f_prime(x_vals[-1])
    x_vals.append(x_new)

plt.plot(x_vals)
plt.show()
```
