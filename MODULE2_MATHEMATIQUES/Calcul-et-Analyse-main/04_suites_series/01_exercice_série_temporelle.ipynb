## TP : Création d'une série temporelle sans bibliothèque avancée

En Data Science, les séries temporelles comptent parmi les modeles les plus courant. Il existe de nombreuses librairies permettant de générer de tels modeles, mais ici nous allons retourner aux bases, avec le modele d'**Auto-Regression**, qui n'est autre qu'un modele linéaire exprimé sous forme de **suite** 

$$ X_{t+1} = \phi X_{t} + \varepsilon_t$$

ici : 
- $X_{t}$ est la valeur de X au temps t
- $X_{t+1}$ est la valeur de X au temps t+1
- $\phi$ est un coefficient
- $\varepsilon_t$ est un *bruit*, c'est-a-dire une valeur aléatoire qui "brouille" le signal

### Expérience : lorsque $\Phi$ est < 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
phi = 0.9
n = 20  # nombre de points
np.random.seed(42)
```

```python
# Initialisation de la suite
X = np.zeros(n)
X[0] = 3

# Génération de la suite
for t in range(1, n):
    X[t] = phi * X[t-1]


plt.figure(figsize=(12, 5))
plt.plot(X, label="Série AR(1)")
plt.title("Simulation d'une série AR(1)")
plt.xlabel("Temps")
plt.ylabel("Valeur")
plt.grid(True)
plt.legend()
plt.show()
```

### Lorsqu'on rajoute $\varepsilon_t$

```python
phi = 0.7
n = 200  # nombre de points
np.random.seed(42)
epsilon = np.random.normal(0, 1, n)
```

```python
# Initialisation de la suite
X = np.zeros(n)
X[0] = 3

# Génération de la suite
for t in range(1, n):
    X[t] = phi * X[t-1]  + epsilon[t]


plt.figure(figsize=(12, 5))
plt.plot(X, label="Série AR(1)")
plt.title("Simulation d'une série AR(1)")
plt.xlabel("Temps")
plt.ylabel("Valeur")
plt.grid(True)
plt.legend()
plt.show()
```

### Modele à plusieurs facteurs

dans la pratique, $X_t$ peut dépendre de plusieurs valeurs de $X$ passées.

Par exemple, le prix du blé à un mois donné $X_t$ peut dépendre du prix du mois précédent $X_{t-1}$ et celui d'il y a 2 mois $X_{t-2}$.

On peut ainsi définir un modele d'Autorégression à 2 facteurs, que l'on note AR(2)

- Modele AR(1) : $ X_t = \phi_1 X_{t-1} +  \varepsilon_t$

- Modele AR(2) : $ X_t = \phi_1 X_{t-1} + \phi_2 X_{t-2} + \varepsilon_t$

- Modele AR(N) : $ X_t = \phi_1 X_{t-1} + \phi_2 X_{t-2} + \cdots + \phi_p X_{t-p} + \varepsilon_t$


```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres du modèle AR(2)
phi = [0.6, -0.4]
n = 200  # nombre de points
np.random.seed(42)
epsilon = np.random.normal(0, 1, n)

# Initialisation de la série
X = np.zeros(n)

# Génération de la série AR(3)
for t in range(3, n):
    X[t] = phi[0]*X[t-1] + phi[1]*X[t-2]  + epsilon[t]

# Tracé de la série
plt.figure(figsize=(12, 5))
plt.plot(X)
plt.title("Simulation d'une série AR(2)")
plt.xlabel("Temps")
plt.ylabel("Valeur")
plt.grid(True)
plt.legend()
plt.show()
```
