# Séries de Taylor


## Introduction

Une **série de Taylor** est un développement en série d'une fonction suffisamment dérivable autour d'un point $a$.  
Elle permet d'approximer la fonction par un polynôme de degré fini : le **polynôme de Taylor d'ordre $n$**.

$$
f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!}\,(x-a)^n
$$

### À quoi ça sert ?
* Approximations numériques rapides (calculs embarqués, calculatrices, GPU, etc.)
* Évaluation de limites et étude locale de fonctions en analyse
* Linéarisation de fonctions non linéaires (contrôle, optimisation, séries temporelles)
* Méthodes de résolution d’équations différentielles



## Séries de Taylor usuelles (développées en $0$)

| Fonction | Série de Taylor | Domaine de convergence |
|----------|-----------------|------------------------|
| $e^{x}$ | $\displaystyle \sum_{n=0}^{\infty} \frac{x^{n}}{n!}$ | $\mathbb R$ |
| $\sin x$ | $\displaystyle \sum_{k=0}^{\infty} (-1)^{k}\frac{x^{2k+1}}{(2k+1)!}$ | $\mathbb R$ |
| $\cos x$ | $\displaystyle \sum_{k=0}^{\infty} (-1)^{k}\frac{x^{2k}}{(2k)!}$ | $\mathbb R$ |
| $\ln(1+x)$ | $\displaystyle \sum_{n=1}^{\infty} (-1)^{n+1}\frac{x^{n}}{n}$ | $-1 < x \le 1$ |
| $(1+x)^{\alpha}$ | $\displaystyle \sum_{n=0}^{\infty} {\alpha \choose n}x^{n}$ | $|x|<1$ |



```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python

def taylor_cos(x, N):
    """Polynôme de Taylor d'ordre N (parité respectée) pour cos(x) centré en 0."""
    series = np.zeros_like(x, dtype=float)
    for k in range(N + 1):
        series += ((-1)**k) * x**(2*k) / math.factorial(2*k)
    return series
```

```python

# Domaine de tracé
x = np.linspace(-4*np.pi, 4*np.pi, 400)
true_values = np.cos(x)

# Différents ordres de développement
orders = [0, 2, 4, 6, 10]

for N in orders:
    plt.figure(figsize=(8, 5))
    plt.plot(x, true_values, label='cos(x)', linewidth=2)
    plt.plot(x, taylor_cos(x, N), linestyle='--', label=f'N = {N}')
    plt.title('Approximation de cos(x) par polynômes de Taylor')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.ylim(-1.5, 1.5)
    plt.show()
```

On remarque que plus l'ordre de la série est élevé, plus l'approximation est précise!
